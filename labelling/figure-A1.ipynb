```python
def plot_alphas(a, alphas_dict, new_1, new_2, category):
    # plots alpha values along one labeling dimension over all 14 batches
    median_alphas = []
    for i in alphas_dict:
        for j in alphas_dict[i]:
            median_alphas.append(j[0])
    median_alphas = np.mean(median_alphas)

    a.set_ylim([-0.1, 1.1])
    a.set_xticks(list(range(1, 15)))
    for i in alphas_dict:
        for j in alphas_dict[i]:
            if j[1] == 'LT-AS' or j[1] == 'AS-LT':
                a.scatter(i, j[0], c='#EB5353', s=20)
            elif j[1] == 'LT-AH' or j[1] == 'AH-LT':
                a.scatter(i, j[0], c='#F9D923', s=20)
            elif j[1] == 'AS-AH' or j[1] == 'AH-AS':
                a.scatter(i, j[0], c='#36AE7C', s=20)
            elif j[1] == 'LT-EM' or j[1] == 'EM-LT':
                a.scatter(i, j[0], c='#187498', s=20)
            elif j[1] == 'EM-AH':
                a.scatter(i, j[0], c='#791E94', s=20)
            else:
                a.scatter(i, j[0], c='grey', s=20)
    # plot points with AS's new labels in batch 4
    a.scatter(new_1[0], new_1[1], c='#36AE7C', marker='^', s=20)
    a.scatter(new_2[0], new_2[1], c='#EB5353', marker='^', s=20)
    a.axhline(y=median_alphas, c='lightgrey', linestyle='--', linewidth=1)
    a.set_xlabel('batch')
    a.set_ylabel('alpha')
    a.set_title(f"{category}")
```

```python
fig, axs = plt.subplots(3, 2, figsize=(14, 14), tight_layout=True)
plot_alphas(axs[0, 0], hate_alphas, (4, 0.51), (4, 0.63), 'HATE')
plot_alphas(axs[0, 1], target_alphas, (4, 0.60), (4, 0.72), 'TARGET')
plot_alphas(axs[1, 0], group_alphas, (4, 0.46), (4, 0.39), 'GROUP')
plot_alphas(axs[1, 1], goal_alphas, (4, 0.50), (4, 0.46), 'GOAL')
plot_alphas(axs[2, 0], strategy_alphas, (4, 0.30), (4, 0.45), 'STRATEGY')
axs[2, 1].axis('off')

legend_elements = [Line2D([0], [0], marker='o', color='w', label='LT-AS', markerfacecolor='#EB5353', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='LT-AH', markerfacecolor='#F9D923', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='AS-AH', markerfacecolor='#36AE7C', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='LT-EM', markerfacecolor='#187498', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='EM-AH', markerfacecolor='#791E94', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='validation', markerfacecolor='grey', markersize=8),
                   Line2D([0], [0], marker='^', color='w', label='re-labeled', markerfacecolor='white',
                          markeredgecolor='grey', markersize=8),
                   Line2D([0], [0], color='lightgrey', label='median alpha', linestyle='--', linewidth=1)]
plt.legend(handles=legend_elements)

plt.show()
# plt.savefig(f'alphas_summary.pdf')
```
